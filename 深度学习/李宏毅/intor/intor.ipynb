机器学习预测油管收益的三步骤：
1. 写出一个具有未知参数的函数 $y = b + wx_1$
2. 定义一个损失函数 define loss from training data   
         -- loss is a function of parameter $L(b,w)$  
         -- loss: $L = \frac {1} {N} \sum_1^N e_n $    
         -- $e = \lvert y -\hat y \rvert$   
            　　---L is mean absolute error MAE  
         -- $e = ( y - y^2 )$  
            　　---L is mean square error MSE
3. 最优化 optimization  
         -- $w^*,b^* = arg\ min_{w,b} \ L $  
         　　--- gradient descent（存在局部最优解的问题但不是真正的痛点）
